```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python

adj_ccc=np.loadtxt('AA.S0044.BXZ_CCC.adj')
adj_cc=np.loadtxt('AA.S0044.BXZ_CC.adj')
adj_wf=np.loadtxt('AA.S0044.BXZ_wm.adj')
adj_mt=np.loadtxt('AA.S0044.BXZ_mt.adj')
adj_ft=np.loadtxt('AA.S0044.BXZ_tf.adj')
misfitname=['CC time shift','multitaper time shift','waveform L2',\
           'TF phase shift','CCC misfit']
wins=[263-12,293-12]
plt.figure(figsize=[10, 8])

waveform=np.loadtxt('EVENT1/REF_SEM/AA.S0044.BXZ.semd')
waveform_smooth=np.loadtxt('EVENT1/REF_SEM_smooth/AA.S0044.BXZ.semd')
plt.subplot(211)
plt.plot(waveform.T[0],waveform.T[1]/np.max(waveform.T[1]),'b',label='anomaly')
plt.plot(waveform_smooth.T[0],waveform_smooth.T[1]/np.max(waveform_smooth.T[1]),'k',label='anomaly_smooth')
plt.xlim(225,350)
plt.ylim(-0.3,0.3)
re = patches.Rectangle((wins[0], plt.ylim()[0]),
                               wins[1] - wins[0],
                               plt.ylim()[1] - plt.ylim()[0],
                               color="lightblue", alpha=0.6)
plt.gca().add_patch(re)
plt.legend()
plt.title('Waveform (receiver 44)')

plt.subplot(212)

#plt.plot(adj_ccc.T[0],adj_cc.T[1]/np.max(adj_cc.T[1]),'k',label='CC time shift')
plt.plot(adj_ccc.T[0],adj_wf.T[1]/np.max(adj_wf.T[1]),'grey',label='waveform L2')

#plt.plot(adj_ccc.T[0],adj_mt.T[1]/np.max(adj_mt.T[1]),'lightblue',label='multitaper time shift')
plt.plot(adj_ccc.T[0],adj_ccc.T[1]/np.max(adj_ccc.T[1]),'b',label='CCC misfit')
#plt.plot(adj_ccc.T[0],adj_ft.T[1]/np.max(adj_ft.T[1]),'red',label='TF phase shift', linewidth=0.5)
plt.xlim(225,350)
plt.ylim(-1.2,1.2)
re = patches.Rectangle((wins[0], plt.ylim()[0]),
                               wins[1] - wins[0],
                               plt.ylim()[1] - plt.ylim()[0],
                               color="lightblue", alpha=0.6)
#plt.gca().add_patch(re)
plt.legend()
plt.title('Adjoint Sources (No time reversal)')
plt.savefig('44_waveform_adjoint.pdf')
```
